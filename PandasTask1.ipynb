# chicago_food_inspection.csv
strip whitespaces and change the case to lowercase for when the Risk is Risk 1(Do it with string function and then with apply function)

```python
df_food = pd.read_csv('D:/Gritfeat/Python/data/chicago_food_inspections.csv')

df_food
```

```python
def clean_name(risk, name):
    return name.strip().lower() if risk == 'Risk 1 (High)' else name


df_food['Name'] = df_food.apply(lambda row: clean_name(row['Risk'], row['Name']), axis=1)

df_food
```
